####**1.1. INTRODUÇÃO**

<p align = "justify">Um dos objetivos de uma biblioteca é promover o uso da informação. Como escreveu Ranganathan (considerado o pai da biblioteconomia) em sua primeira lei: “Os livros são para serem usados”. Por isso, o empréstimo dos materiais em uma biblioteca é uma das formas de se indicar o uso daquela informação.

<p align = "justify">Os empréstimos realizados podem ser usados com um indicador, mesmo que de forma básica (pois você não consegue garantir que haja uma leitura ou utilização real), por este motivo, analisar a quantidade de empréstimos se torna importante para entender a dinâmica do funcionamento de uma biblioteca.

<p align = "justify">As instituições de ensino superior (IES) têm a necessidade de passar por avaliações do Ministério da Educação (MEC) para que possam ofertar e continuar ofertando cursos de graduação e pós-graduação. A biblioteca universitária faz parte de um dos indicadores da avaliação dos cursos, em principalmente três aspectos: acervo, infraestrutura e serviços. Dentre os serviços avaliados, a presença de indicadores sobre o uso dos materiais do acervo (empréstimos, consultas) é um deles.

<p align = "justify">Assim, desenvolver estratégias para o uso da informação é uma tarefa constante, devido às diversas mudanças que ocorrem o tempo todo: os usuários estão evoluindo seus conhecimentos, novos alunos chegam, alunos que se formam saem e as informações estão sempre em movimento e transformação.



####**1.2. PROBLEMA**

<p align = "justify"> Uma nova gestão assumiu as atividades da Biblioteca Central da Universidade do Rio Grande do Norte e eles gostariam de entender o como está o funcionamento da biblioteca. Esta gestão será a primeira a utilizar os dados de empréstimos para traçar os planos e metas para os próximos anos, tanto em estratégias para atrair novos usuários quanto para a parte organizacional e manutenção dos atuais acervos. Assim, eles precisam compreender melhor como as atividades dentro da biblioteca estão sendo realizadas e para isso, analisarão os dados para saber quais acervos necessitam de constantes investimentos em manutenção, aquisição de novos materiais, quais acervos necessitam de investimento em marketing para atrair mais público ou até mesmo, repassar as obras que não estão sendo utilizadas, uma vez que para mantê-las é necessário uma série de investimentos em infraestrutura adequada.

####**1.3. OBJETIVOS**

<p align = "justify">Com base nas necessidades da diretoria da biblioteca, serão abordados três pontos principais,<b>1)</b> evidenciar a tendência de empréstimos geral da biblioteca; <b>2)</b> descobrir quais acervos são os mais acessados e quais cursos mais utilizam o serviço de empréstimos da biblioteca e <b>3)</b> com foco mais organizacional, ver  quais são os meses de maior e menor movimento da biblioteca.


####**1.4. PERGUNTAS**

<p align = "justify">Após traçar os objetivos, as perguntas a seguir servirão de orientação para a realização da EDA (<i>Exploratory Data Analysis</i>):
<b>1)</b> Qual a tendência de empréstimo de livros entre os anos? <b>2)</b> Na presença de uma tendência, há diferenças entre os anos? <b>3)</b> Entre os meses de cada ano, há diferença no número de empréstimos? <b>4)</b> Quais são os melhores meses para que os funcionários possam realizar suas férias sem comprometer o funcionamento da biblioteca? <b>5)</b> Qual é o maior horário de pico e o menor dentro de um dia? <b>6)</b> Quais são os assuntos mais procurados pelos estudantes? <b>7)</b> Quais os cursos que mais utilizam o serviço de empréstimo da biblioteca?</p>

<p align = "justify">Somente depois do conhecimento destas outras informações, será possível entender o cenário e apresentá-lo à diretoria da biblioteca para que possam tomar melhores decisões na melhoria da infraestrutura, dos recursos e processos da unidade de informação, melhorando a assertividade do marketing em atrair mais usuários e adequação dos custos a cada situação.

##6.0. TESTES DE HIPÓTESES



###6.1. Empréstimo - Ano

<p align = "justify">Após a análise dos gráficos de distribuição do nº de empréstimos totais entre os anos, será realizado uma análise com teste de hipótese para evidenciar se há ou não diferença entre os anos amostrados. Para que seja escolhido o teste adequado, será realizado o teste de normalidade dos dados.

####6.1.1 Teste de Normalidade

```python
from scipy.stats import normaltest
```

```python
testes_emprestimo_ano = df_emprestimo_ano.iloc[:,1:11]
stat, p_value = normaltest(testes_emprestimo_ano)
print(f'Estatística de teste: {stat}, Valor-p dos grupos: {p_value}')
```

<p align = "justify">Como a matriz de dados não apresenta nº amostral suficiente para a realização do teste de normalidade (normaltest), será realizado outro teste de normalidade adequado para amostras com n menor que 20. O teste utilizado será o Shapiro.

```python
from scipy.stats import shapiro
```

```python
stat, p_value = shapiro(testes_emprestimo_ano)
print(f'Estatística de teste: {stat}, Valor-p: {p_value}')
```

```python
alpha = 0.05
if p_value > alpha:
    print('Os dados parecem seguir uma distribuição normal.')
else:
    print('Os dados não seguem uma distribuição normal.')
```

Como os valores de empréstimos anuais não seguem uma distribuição normal, será realizado uma normalização logarítmica a fim de verificar se os dados apresentam uma distribuição normal.

```python
testes_emprestimo_ano_transformados = np.log(testes_emprestimo_ano)
```

```python
stat, p_value = shapiro(testes_emprestimo_ano_transformados)
print(f'Estatística de teste: {stat}, Valor-p: {p_value}')
```

```python
if p_value > alpha:
    print('Os dados parecem seguir uma distribuição normal.')
else:
    print('Os dados não seguem uma distribuição normal.')
```

####6.1.2 Teste de Hipóteses - KRUSKAL-WALLIS

<p align = "justify">Novamente os valores de empréstimo anual não apresentam distribuição normal, mesmo depois de ter sido realizada uma normalização. Assim, para a realização do teste de hipótese será utilizado o teste de Kruskal-Wallis.

Onde a hipótese nula representa que os grupos testados(anos) apresentam diferenças significativas. Já a hipótese alternativa é de os grupos não variam.

```python
from scipy.stats import kruskal
stat, p_value = kruskal(ano_2010, ano_2011, ano_2012, ano_2013,
                        ano_2014, ano_2015, ano_2016, ano_2017,
                        ano_2018, ano_2019)
print(f'Estatística de teste: {stat}, Valor-p: {p_value}')
```

```python
if p_value > alpha:
    print('Aceita-se a hipótese nula e os dados não mostraram diferenças significativas no nº de empréstimos entre os anos.')
else:
    print('Rejeita-se a hipótese nula e os dados mostraram diferenças significativas no nº de empréstimos entre os anos.')
```

<p align = "justify">Apesar dos empréstimos anuais terem sofrido flutuações entre os anos, principalmente nos anos de 2018 e 2019 que mostraram tendência de queda, não foi verificado diferença significativa entre os anos, sendo assim, assume-se que as variações são naturais dentro do período amostrado.

No entanto, devido às grandes mudanças tecnológicas e políticas dentro desse período de 10 anos, a variação desse período pode mascarar mudanças iniciadas mais recentemente, como a chegada de e-readers ao mercado brasileiro e sua constante crescente popularização, é um dos exemplos que podem estar sendo mascarado. Assim, será performado um novo teste hipótese com um período de tempo menor, de 2015-2019, já que o ano de 2020 está excluído.

```python
stat, p_value = kruskal(ano_2015, ano_2016, ano_2017,
                        ano_2018, ano_2019)
print(f'Estatística de teste: {stat}, Valor-p: {p_value}')
```

```python
if p_value > alpha:
    print('Aceita-se a hipótese nula e os dados não mostraram diferenças significativas no nº de empréstimos entre os anos.')
else:
    print('Rejeita-se a hipótese nula e os dados mostraram diferenças significativas no nº de empréstimos entre os anos.')
```

Novamente, não tivemos diferenças entre os empréstimos anuais, sendo assim, podemos assumir com tranquilidade que não diferenças significativas entre os anos mais recentes amostrados.
